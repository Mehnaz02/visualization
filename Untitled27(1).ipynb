```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
data=data.data
```

```python
data
```

```python
# we statndarlizing
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
ST=sc.fit_transform(data)
```

```python
#now performing Agglomerative Clustering model based on the data

from sklearn.cluster import AgglomerativeClustering
```

```python
AG=AgglomerativeClustering(n_clusters=3,linkage='single')
```

```python
cluster=AG.fit(ST)
```

```python
Clusters=cluster.labels_
```

```python
# ploting the dendogram
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
la=linkage(ST,method='single')
```

```python
la
```

```python
plt.figure(figsize=(17,5))
dendrogram(la)
plt.show()
```
